```r
library(MASS); library(tree)
pima.tree = tree(type ~ ., data = Pima.tr)
plot(pima.tree); text(pima.tree, pretty = 0)
summary(pima.tree)
```

```r
cv.pima=cv.tree(pima.tree, )

plot(cv.pima$size ,cv.pima$dev ,type="b")
pima.prune = prune.tree(pima.tree, best = 4)
plot(pima.prune); text(pima.prune, pretty = 0)
```

```r
pima.pred = predict(pima.prune, Pima.te, type = "class")
table(pima.pred, Pima.te$type)
```
