```r
library(foreign)
library(Matching)
data(lalonde, package = "Matching")

df = read.dta("dw_data.dta")
```

```r
set.seed(931139)

psmodel = glm(treat ~ age + education + black + hispanic + married + nodegree + re74 + re75, family = binomial(), data = df)
psmatch = Match(Tr = df$treat,M = 1,X = psmodel$fitted.values, replace = FALSE, caliper = NULL)

# matched = lalonde[unlist(psmatch[c("index.treated","index.control")]), ]
```

```r
mb  = MatchBalance(treat ~ age + education + black + hispanic + married + nodegree + re74 + re75, data=df, match.out=psmatch, nboots=10)
```

```r
genout = GenMatch(Tr = df$treat,M = 1,X = psmodel$fitted.values)
mout = Match(Tr = df$treat,M = 1,X = psmodel$fitted.values, estimand = "ATE", Weight.matrix = genout)
summary(mout)
```
